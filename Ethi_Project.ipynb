```python
for folder in os.listdir('facebook-sampledataset'):
    if folder == '.DS_Store' or folder == '':
        continue
    print(folder + '\n')
    for file in os.listdir(os.path.join('facebook-sampledataset', folder)):
        if file.endswith('json'):
            path = os.path.join('facebook-sampledataset', folder, file)
            f = open(path, 'r')
            print(json.loads(f.read()))
        print('\n')
    #print(os.path.join('facebook-sampledataset', filename))
```
