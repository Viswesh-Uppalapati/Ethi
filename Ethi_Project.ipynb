```python
import pandas as pd
import numpy as np
import os
import json
```

#### Advertisement Data

```python
# ad_interests containst all the categories of ads that the user has an interes in

ad_interests_path = os.path.join('facebook-sampledataset', 'ads_and_businesses', 'ads_interests.json')
ad_interests = pd.Series(json.loads(open(ad_interests_path, 'r').read())['topics'])
ad_interests
```

```python
ad_interacted_path = os.path.join('facebook-sampledataset', 'ads_and_businesses', 
                                  'advertisers_you\'ve_interacted_with.json')
ads_interacted_dict = json.loads(open(ad_interacted_path, 'r').read())['history']
```

```python
# ads_interacted_with is a table that contains the ads that the user clicked on

ads_interacted_with = {'title': [], 'action': [], 'timestamp': []}
for x in ads_interacted_dict:
    ads_interacted_with['title'].append(x['title'])
    ads_interacted_with['action'].append(x['action'])
    ads_interacted_with['timestamp'].append(x['timestamp'])

ads_interacted_with = pd.DataFrame(ads_interacted_with)
ads_interacted_with.head()
```

```python
off_fb_path = os.path.join('facebook-sampledataset', 'ads_and_businesses', 
                                  'your_off-facebook_activity.json')
off_fb_dict = json.loads(open(off_fb_path, 'r').read())['off_facebook_activity']
```

```python
# off_fb_activity contains all the pages the user visited outside of facebook

off_fb_activity = {'website': [], 'page_id': [], 'activity_type': [], 'timestamp': []}
for x in off_fb_dict:
    for y in x['events']:
        off_fb_activity['website'].append(x['name'])
        off_fb_activity['page_id'].append(y['id'])
        off_fb_activity['activity_type'].append(y['type'])
        off_fb_activity['timestamp'].append(y['timestamp'])

off_fb_activity = pd.DataFrame(off_fb_activity)
off_fb_activity
```

#### Comments Data

```python
comments_path = os.path.join('facebook-sampledataset', 'comments', 'comments.json')
comments_dict = json.loads(open(comments_path, 'r').read())['comments']
```

```python
# comments table contains all the user comments data. There are some null values in the comment_content and group_posted_in
# columns.

comments = {'timestamp': [], 'comment_content': [], 'group_posted_in': [], 'action': []}
for x in comments_dict:
    comments['timestamp'].append(x['timestamp'])
    comments['action'].append(x['title'])
    try:
        comments['comment_content'].append(x['data'][0]['comment']['comment'])
    except:
        comments['comment_content'].append(np.nan)
    try:
        comments['group_posted_in'].append(x['data'][0]['comment']['group'])
    except:
        comments['group_posted_in'].append(np.nan)

comments = pd.DataFrame(comments)
comments
```

#### Likes and Reactions

```python
pages_liked_path = os.path.join('facebook-sampledataset', 'likes_and_reactions', 'pages.json')
pages_liked_dict = json.loads(open(pages_liked_path, 'r').read())['page_likes']
```

```python
# pages_liked table has all the pages that the user has liked

pages_liked = {'page_name': [], 'timestamp': []}
for x in pages_liked_dict:
    pages_liked['page_name'].append(x['name'])
    pages_liked['timestamp'].append(x['timestamp'])

pages_liked = pd.DataFrame(pages_liked)
pages_liked
```

```python
posts_liked_path = os.path.join('facebook-sampledataset', 'likes_and_reactions', 'posts_and_comments.json')
posts_liked_dict = json.loads(open(posts_liked_path, 'r').read())['reactions']
```

```python
# posts_liked has all user interactions and reactions to a multitude of different posts and comments on facebook

posts_liked = {'timestamp': [], 'reaction_type': [], 'action_performed': []}
for x in posts_liked_dict:
    posts_liked['timestamp'].append(x['timestamp'])
    posts_liked['reaction_type'].append(x['data'][0]['reaction']['reaction'])
    posts_liked['action_performed'].append(x['title'])

posts_liked = pd.DataFrame(posts_liked)
posts_liked
```

#### User Messages

```python
# All the message data from every chat of the user. Note: The user name in sample dataset is 'ACWBUYF.XYFFS' and some of
# the data from the 'message' column is missing.

messages = {'inbox_name': [], 'participants': [], 'message': [], 'message_type': [], 'sender': [], 'timestamp': []}

for folder in os.listdir(os.path.join('facebook-sampledataset', 'messages', 'inbox')):
    message_path = os.path.join('facebook-sampledataset', 'messages', 'inbox', folder, 'message_1.json')
    
    if folder == '.DS_Store':
        continue
        
    current_dict = json.loads(open(message_path, 'r').read())
    
    for x in current_dict['messages']:
        messages['inbox_name'].append(current_dict['title'])
        
        participants = []
        for y in current_dict['participants']:
            participants.append(y['name'])
            
        messages['participants'].append(participants)
        
        try:
            messages['message'].append(x['content'])
        except:
            messages['message'].append(np.nan)
            
        messages['sender'].append(x['sender_name'])
        messages['message_type'].append(x['type'])
        messages['timestamp'].append(x['timestamp_ms'])

messages = pd.DataFrame(messages)
messages
```
